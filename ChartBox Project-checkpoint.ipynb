```python
import yfinance as yf
import pandas as pd
import numpy as np
```

```python
data = yf.download("AAPL", start="2022-10-18", end="2023-10-18")
print(data)
```

```python
df = pd.DataFrame(data)
df.head()
```

```python
df.info()
```

```python
df["Open-Close(Candle-Body)"] = df["Close"] - df["Open"]
```

```python
df.head()
```

```python


df['Candle'] = ['Bearish' if x < 0 else 'Bullish' for x in df["Open-Close(Candle-Body)"]]
```
